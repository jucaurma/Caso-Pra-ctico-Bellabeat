# Introducción:

Entre sus productos cuenta con:

## Observaciones:


* Vemos que la mayoría de usuarias está en 10.000 pasos y una cantidad menor está en torno a los 20.000. La Organización Mundial de la salud recomienda que los adultos de 18 a 64 años deben realizar al menos 150 minutos de actividad física aeróbica moderada o 75 minutos de actividad física aeróbica vigorosa a la semana, o una combinación equivalente de ambos. Si bien la Organización Mundial de la Salud no establece un objetivo específico en términos de pasos diarios, caminar puede ser una forma efectiva de cumplir con las recomendaciones de actividad física, especialmente si se camina a un ritmo moderado o vigoroso durante períodos prolongados. La meta comúnmente citada de 10,000 pasos al día es una forma de fomentar la actividad física diaria y puede ser consistente con las pautas de la OMS, aunque no es una recomendación oficial de la organización.  



* La mayoría de usuarias recorren una media de 5 Km al día y un porcentaje mucho menor recurre 10 o más Km.  



* En la gráfica correspondiente a los minutos sendentarios observamos dos picos acentuados que representa a dos segmentos, uno que corresponde a un grupo de usuarias que pasan más de 16 horas de forma sedentaria y otro que pasan menos de 16 horas sedentarias. Esto nos puede llevar a pensar que uno de los grupos no registran su actividad porque no realizan actividad física, porque no saben registrarla o porque utilizan el dispositivo con fines meramente estéticos.

* La mayoría de usuarias tienen un consumo medio de calorías en torno a 2.000 calorías/día. Según la Organización Mundial de la Salud las pautas dietéticas pueden sugerir que las mujeres adultas necesitan consumir entre 1,800 y 2,400 calorías por día, por lo que podemos llegar a la idea de que la mayoría de las usuarias tienen un consumo de calorías dentro de la media y por tanto pueden aumentar su actividad física y así tener un consumo calórico por día mayor.

Vamos a ver con un poco más de detalle cómo es la segmentación de las usuarias respecto de las calorías consumidas:

```python
bins = [float('-inf'), 1800, 2400, float('inf')]
labels=['Menos de 1800', 'Entre 1800 y 2400', 'Más de 2400']
df['segmentos_de_calorias'] = pd.cut(df['Calories'], bins = bins, labels = labels, include_lowest=True)

category_porcentajes = df['segmentos_de_calorias'].value_counts(normalize=True) * 100
category_porcentajes_sorted = category_porcentajes.sort_index()

plt.figure(figsize=(10,6))
category_porcentajes_sorted.plot(kind = 'bar', color = 'purple')

plt.title('Segmentos de porcentajes')
plt.xlabel('Segmento de calorías')
plt.ylabel('Porcentaje')
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)

for index, value in enumerate(category_porcentajes_sorted):
    plt.text(index, value + 0.1, f'{value:.2f}%', ha='center', color='black', fontsize=9)

plt.xticks(rotation=0)


plt.tight_layout()
plt.show()

```
